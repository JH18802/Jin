# **희망 직종 체크 및 요구 조건을 통한 데이터 출력**

```python
# 전체 직종 출력

df["HOPE_JSSFC_CMMN_CODE_SE_NM_1"].unique()
```

```python
# 각 직종별 희망 인원

df["HOPE_JSSFC_CMMN_CODE_SE_NM_1"].value_counts()
```

```python
# WORK_AREA_CMMN_CODE_SE_1_1이 서울이 아닌 데이터 출력
# 괄호 안의 내용이 요구조건

df.query(''' WORK_AREA_CMMN_CODE_SE_1_1 != "서울" ''')
```

```python
# ACDMCR_CMMN_CODE_NM이 고등학교인 데이터 출력

df.query(''' ACDMCR_CMMN_CODE_NM == "고등학교" ''')
```

```python
# ACDMCR_CMMN_CODE_NM이 고등학교인 데이터의 개수 확인

highschool = df.query(''' ACDMCR_CMMN_CODE_NM == "고등학교" ''')
highschool.shape[0]
```

```python
# 60세 이상이면서 서울에서 일하는 것을 희망하는 남성들의 데이터 출력

df.query(''' AGE >= 60 and SEX == '남성' and WORK_AREA_CMMN_CODE_SE_1_1 == '서울' ''')
```

```python
# 60세 이하이면서 경기를 희망하지 않는 사람들의 데이터 출력

df.query(''' AGE <= 60 and WORK_AREA_CMMN_CODE_SE_1_1 != '경기' ''')
```

```python
# 10세 이상이면서 희망직종 코드가 80000이상인 사람들의 데이터 출력

df.query(''' AGE >= 10 and HOPE_JSSFC_CMMN_CODE_SE_1 >= 800000 ''')
```

```python
# 서울시 강서구에서 근무를 희망하는 여성의 데이터 출력

df.query(''' SEX == '여' and WORK_AREA_CMMN_CODE_SE_1_1 ==  '서울' and WORK_AREA_CMMN_CODE_SE_1_2 == '강서구' ''')
```

```python
# 희망 직종 출력(재가 요양보호사)

df.query(''' HOPE_JSSFC_CMMN_CODE_SE_NM_1 == "재가 요양보호사" and SEX == "여" ''')
```
